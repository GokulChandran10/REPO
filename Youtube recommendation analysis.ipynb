# Data Pre-processing 

### Checking For Null Values and Duplicates

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
print("shape of the dataset after removing the duplicates:", df.shape)
```

**_Shows there is no null values and duplicates._**

### Creating New Coloumns

```python
df['video_length_type']=np.where(df['duration_seconds'] <= 600, 'Short','Long')
df.head()
```

**_This shows the how long the video is (Categorised based on duration) and this helps in understanding how the videos perform based on their length._** 

```python
df['engagement_level']=pd.cut(df['engagement_rate'], bins=[0,0.01,0.03,1],labels=['Low','Medium','High'])
df.head()
```

**_This show how hooked people are to certain youtube videos. The level of engagement varies from low to High with Medium level of engagement in between. Engagement rate here is calucalted by dividing the sum of like and comment count with view count._**

```python
df['video_age_category'] = pd.cut(
    df['video_age_days'],
    bins=[-1,30,365,2000,df['video_age_days'].max()],
    labels=['new','recent','old','very old']
)
df.head()
```

**_Based on how old the video(How many days it has been since it is uploaded) is we can categorize them from new to very old._** 

```python
df['view_category'] = pd.cut(
    df['view_count'],
    bins=[0, 1e6, 1e7, 1e8, df['view_count'].max()],
    labels=['Low', 'Medium', 'High', 'Viral']
)
df.head()
```

***View Category Shows based on the amount of view, how well the video performed and which categrory it belongs to. (The category varies from low performance to Viral Video)***

# Exploratory Data Analysis

```python
df.groupby('video_length_type')['engagement_rate'].mean()
```

***Shows longer the video length shorter engagement rate this can often be thought that people might be getting bored halfway through video or maybe even not finishing the entire video***

# UNIVARIATE ANALYSIS

## *Distribution of view count (Based on how much views each video gets)*

```python
plt.style.use('dark_background')
sns.histplot(df['view_count'], bins=40, kde=True, color='cyan')
plt.title("Distribution of Video Views", fontweight='bold')
plt.xlabel("View Count")
plt.ylabel("Frequency")
plt.show()
```

### INSIGHT: 
#### The distribution of video count shows that most videos receive relatively low views, while a small number of videos achieve extremely high(viral) view counts.

## *Videos by View Category*

```python
df['view_category'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['gold', 'coral', 'skyblue', 'cyan'])
plt.title("Distribution of Videos by View Category", fontweight='bold')
plt.ylabel("")
plt.show()
```

### INSIGHT: 
#### In accordance with the above histogram this proves that Viral videos get more percentage of views but is less in numbers.

## *Videos By View Category*

```python
df['view_category'].value_counts().plot(kind='bar', color='cyan')
plt.title("Number of Videos by View Category",fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

### INSIGHT:
####  Majority of videos falls under Medium and High performance but very few reach Viral stage.

## *Distribution of engagement rate*

```python
plt.figure()
plt.hist(df['engagement_rate'], bins=40, color='cyan')
plt.title("Distribution of Engagement Rate",fontweight='bold')
plt.xlabel("Engagement Rate")
plt.ylabel("Frequency")
plt.show()
```

### INSIGHT:
#### Most videos have low engagemnet rates with very few videos having high level of engagement.

## *Videos by engagement level*

```python
df['engagement_level'].value_counts().plot(kind='bar',color='cyan')
plt.title("Videos by Engagement Level", fontweight='bold')
plt.xlabel("Engagement Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

### INSIGHT:
#### Most Videos have medium level engagement which shows content value is still very much in the balance even though very few have high engagemnet most of the videos gets decent amount of views.

# BIVARIATE ANALYSIS

## *Average Views by Video Length*

```python
plt.figure()
df.groupby('video_length_type')['view_count'].mean().plot(kind='bar',color='cyan')
plt.title("Average Views by Video Length", fontweight='bold')
plt.xlabel("Length of the Video")
plt.ylabel("View Count")
plt.xticks(rotation=45)
plt.show()
```

### INSIGHT
#### Short Videos receive more views on average compared to long videos

## *Duration vs Views*

```python
plt.scatter(df['duration_seconds'], df['view_count'],alpha=0.5,color='cyan')
plt.title("Duration vs Views", fontweight='bold')
plt.xlabel("Duration")
plt.ylabel("Views")
plt.show()
```

### INSIGHT
#### There is no strong relation between duration and views. Viral videos can both short and long. Sometimes it might also depend on the content creator

## *Views vs Likes*

```python
plt.figure()
plt.scatter(df['view_count'], df['like_count'], color='cyan',alpha=0.6)
plt.title("Views vs Likes", fontweight='bold')
plt.xlabel("View Count")
plt.ylabel("Like Count")
plt.show()
```

### INSIGHTS
#### This shows the relationship between views and likes. From the graph it is very evident that more views brings more likes indicating a positive relationship between the two. This supports the idea that likes depends on exposure but content quality matters too.

## *Engagement Rate by Video age*

```python
df.groupby('video_age_category')['engagement_rate'].mean().plot(marker='o',color='cyan')
plt.title("Engagement Rate by Video Age Group", fontweight='bold')
plt.xlabel("Age Category")
plt.ylabel("Rate of engagement")
plt.xticks(rotation=45)
plt.show()
```

### INSIGHT
#### New videos tend to have higher engagement, which decreases as the videos become older.

## *Engegement Rate by Video Length*

```python
sns.boxplot(x='video_length_type', y='engagement_rate', data=df, color='cyan')
plt.title("Engagement Rate by Video Length", fontweight='bold')
plt.xlabel("Length of the video")
plt.ylabel("Rate of engagement")
plt.xticks(rotation=45)
plt.show()
```

### INSIGHTS
#### Along with the information gathered from above graph new and short video shows more rate of engagemnet than long videos even though it is not by much this could mean that in the future we might see more youtube shorts than full length videos

## *Engagement Rate based on Engagement Level*

```python
df.groupby('engagement_level')['engagement_rate'].mean().plot(kind='bar',color=['cyan','lime','purple'])
plt.title("Average Engagement Rate by Engagement Level", fontweight='bold')
plt.xlabel("Engagement Level")
plt.ylabel("Engagement Rate")
plt.xticks(rotation=45)
plt.show()
```

### INSIGHT
#### Engagement rate is a numeric measure whereas engagement level is grouped category based on engagement rate. Basically higher the engagemnet level more chances that the video gets more likes and views per views.

## *Views vs Likes and Views vs Comments*

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].scatter(df['view_count'], df['like_count'], color='cyan')
axes[0].set_title("Views vs Likes", fontweight = 'bold')
axes[0].set_xlabel("View Count")
axes[0].set_ylabel("Like Count")

axes[1].scatter(df['view_count'], df['comment_count'], color='magenta')
axes[1].set_title("Views vs Comments", fontweight = 'bold')
axes[1].set_xlabel("View Count")
axes[1].set_ylabel("Comment Count")

plt.tight_layout()
plt.show()
```

### INSIGHT:
#### As the number of views increases likes and comments also increases but engagement varies across videos, some videos have more likes and comments than others even with similar amount of views

# MULTIVARIATE ANALYSIS

## *Correlation Heatmap*

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only = True), cmap='viridis', annot = True)
plt.title("Correlation Heatmap", fontweight='bold')
plt.show()
```

### INSIGHTS:
#### Views and likes have strong coorelation.
 
#### Engagement rate depends on multiple factors.
 
#### Some variables have weak or no relationship at all.

## *Engagement Rate by Length and Level*

```python
pivot_table = pd.pivot_table(
    df,
    values='engagement_rate',
    index='video_length_type',
    columns='engagement_level',
    aggfunc='mean'
)

sns.heatmap(pivot_table, annot=True, cmap='viridis')
plt.title("Engagement Rate by length and Level", fontweight='bold')
plt.show()
```

### INSIGHT 
#### This heatmap shows that videos with higher engagement level has higher engagement rates. Additionally short videos tends to have slighlty higher engagement rates compared to long videos. This shows that shorter content may perform better in terms of audience interaction.
